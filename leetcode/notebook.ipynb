```python
from collections import defaultdict
nums = [1,5,5,2,5,3,5,4]
cnt = defaultdict(int)
for x in nums:
    cnt[x] += 1
print(cnt)
```

```python
from typing import List
def is_no_decreasing(a: List[int]) -> bool:
    for i in range(1, len(a) - 1):
        if a[i] < a[i-1]:
            return False
    return True

nums = [1,2,3,3]
print(is_no_decreasing(nums))
```
